```python
import json
import requests
import pandas as pd
import matplotlib as plt
```

```python
#establish endpoint 
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json'

# 1.Make an API request that returns the months where "fire" was searched in 2016. Which month had the most searches?
params = {'year':'2016', 'query_text':'fire'}
response = requests.get(endpoint, params = params)
```

```python
#check for good response
response
```

```python
#review response
response.text
```

```python
#make it a dictionary
res = response.json()
res
```

```python
#use index to get inside of [] then shows the available keys
res[0].keys()
```

```python
#use a loop to find the highest query count of fire mentions
max_query_count = 0
max_query_month = ''

for item in res:
    query_count = int(item['query_count'])
    if query_count > max_query_count:
        max_query_count = query_count
        max_query_count_month = item['month_name']

max_query_count_month
```

```python
# 2. Make an API request that returns all the times a query was run more than 100 times in a month. How many times did this occur?
params2 = {'$where': 'query_count >= 100'}
response = requests.get(endpoint, params = params2)
```

```python
#check for good response
response
```

```python
res = response.json()
res
```

```python
#use a loop to find the all the times when a query was run more than 100 times in a month
max_query_count = 100
max_query_month = ''

for item in res:
    query_count = int(item['query_count'])
    if query_count > max_query_count:
        max_query_count = query_count
        if query_count > 100:
            max_query_count_month = item['month_name']

max_query_count
```

```python
# 3. Make another API request that returns all the times "codes" was searched more than 100 times in a month. How many times did this occur?
params3 = {'$where': 'query_count >= 100', 'query_text':'codes'}
response = requests.get(endpoint, params = params3)
```
